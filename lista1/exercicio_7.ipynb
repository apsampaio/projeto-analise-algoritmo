```python
def overlap(r1: dict, r2: dict) -> bool:
    r1w = (r1["x2"] - r1["x1"]) * 0.5  
    r2w = (r2["x2"] - r2["x1"]) * 0.5  

    l = r2["x1"] - r1["x1"]

    gap = l - r1w - r2w

    if gap > 0:
        return False
    return True
##
```

Complexity = $n(1)$

```python

r1 = {"x1": 0, "y1": 0, "x2": 2, "y2": 2}
r2 = {"x1": 2, "y1": 2, "x2": 4, "y2": 4}
print(overlap(r1, r2)) # True

r1 = {"x1": 0, "y1": 0, "x2": 1, "y2": 1}
r2 = {"x1": 2, "y1": 2, "x2": 4, "y2": 4}
print(overlap(r1, r2)) # False
```
